```python
a = housing_dropped.columns.values.tolist()
housing_null = {'Column':a,
'number of null values':housing_dropped.isnull().sum(axis = 0) }
housing_null = pd.DataFrame(housing_null)
housing_null.head(5)
```

<font face="vazir" size=3>

```python
housing_no_missing = housing_dropped.copy()
```

</div>

<font face="vazir" color="#0099cc">
قسمت سوم
</font>
</h2>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در دیتافریم <code>housing_no_missing</code> مقادیر خالی ستون <code>DOM</code> را با مقدار مد این ستون جایگزین کنید.  پس از آن ستون‌هایی که مقدار <code>elevator</code> یا <code>subway</code> آنها خالی است را حذف کنید.
</font>
</p>

```python
#housing_no_missing = None # to do
housing_no_missing.mode(axis='rows', numeric_only=True)
housing_no_missing['DOM'] = housing_no_missing['DOM'].fillna(1)
housing_no_missing = housing_no_missing[housing_no_missing['subway'].notna()]
housing_no_missing = housing_no_missing[housing_no_missing['elevator'].notna()]
housing_no_missing.head(10)
```


<div dir=rtl id="start" style="direction:rtl;line-height:200%;">
	<font face="vazir" size=3>

پس از انجام این عملیات دیگر داده گمشده‌ای نداریم و می‌توانیم به مرحله بعد برویم. در مرحله بعد از دیتافریم <code>housing_no_missing</code> استفاده خواهیم کرد. دقت کنید که هنگام ذخیره فایل بر روی دیسک هم باید انکدری انتخاب کنیم که قادر به شناسایی همه کاراکتر‌های دیتافریم باشد.

</p>

```python
import zipfile
import joblib

joblib.dump(housing_dropped.columns,"housing_dropped_cols") 
joblib.dump(housing_null,"housing_null")
joblib.dump(housing_no_missing['DOM'],"housing_no_missing_DOM")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED
    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = ["housing_dropped_cols","housing_null","housing_no_missing_DOM", "1_fill_missing_values.ipynb"]
compress(file_names)
```
